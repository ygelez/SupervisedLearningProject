```python
import math
def  plot_all_histograms(df, title_prefix = ""):
    num_cols = df.select_dtypes(include=[np.number]).columns
    n_cols = 3
    n_rows = math.ceil(len(num_cols)/3)
    plt.figure(figsize=(5*n_cols, 4*n_rows))
    for i,col in enumerate(num_cols, 1):
        plt.subplot(n_rows, n_cols, i)
        sns.histplot(df[col], kde  =True, bins=50)
        plt.title(f"{title_prefix}{col}")
        plt.xlabel("")
        plt.ylabel("")
    plt.tight_layout()
    plt.show()
```

```python
plot_all_histograms(df,title_prefix= "original-" )
```

```python
df = df.drop(["duration", "campaign", "pdays"], axis=1)
```

```python
X = df.drop("deposit", axis = 1)
y = df["deposit"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15, stratify=y)
```

```python
num_cols = X_train.select_dtypes(include=["int64", "float64"]).columns
cat_col = X_train.select_dtypes(include=["object"]).columns
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
df.head()
```

```python
preprocess = ColumnTransformer(
    transformers=[
        ("num", "passthrough", num_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_col) 
    ]
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = Pipeline(steps=[
    ("prep", preprocess),
    ("model", RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1, max_depth=3))

    
])
```

```python
clf.fit(X_train, y_train)
```

```python
print("Train score:", clf.score(X_train, y_train))
print("Test score:", clf.score(X_test, y_test))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_pred = clf.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
rf = clf.named_steps["model"]
feature_names = clf.named_steps["prep"].get_feature_names_out()
```

```python
importances = pd.DataFrame({
    "feature": feature_names,
    "importance": rf.feature_importances_
}).sort_values("importance", ascending=False)

importances.head(15)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf_lr = Pipeline(steps=[
    ("prep", preprocess),
    ("model", LogisticRegression(max_iter=1000))

    
])
```

```python
clf_lr.fit(X_train, y_train)
```

```python
print("Train:", clf_lr.score(X_train, y_train))
print("Test:", clf_lr.score(X_test, y_test))
y_pred_lr = clf_lr.predict(X_test)
print(confusion_matrix(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr))
```

```python
clf_lr_balanced = Pipeline(steps=[
    ("prep", preprocess),
    ("model", LogisticRegression(
        max_iter=1000,
        class_weight="balanced"
    ))
])

clf_lr_balanced.fit(X_train, y_train)

print("Train:", clf_lr_balanced.score(X_train, y_train))
print("Test:", clf_lr_balanced.score(X_test, y_test))
```

```python
y_pred_bal = clf_lr_balanced.predict(X_test)

print(confusion_matrix(y_test, y_pred_bal))
print(classification_report(y_test, y_pred_bal))
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt
y_proba = clf_lr_balanced.predict_proba(X_test)[:, 1]
auc_score = roc_auc_score(
    y_test.map({"no":0, "yes":1}),
    y_proba
)

print("ROC-AUC:", round(auc_score, 3))
```
